# Handling Missing Values from the dataset

## According to the rules:

a. If there are null values in numerical column, replace the null values by the mean of that column

b. If there are null values in categorical column, replace the null values by the mode of that column

c. If more than 50% the values in a column are null, then drop that entire column

```python
covid_num.isnull().sum()
```

```python
(covid_num.isnull().sum())>(len(covid_num))/2
```

Since all the columns have null values not more than 50% of the total rows, we won't drop any column.

## Predicting the Total Cases on a given Date

To get the prediction, Train-Test-Split does the job of accurately cutting out the rows & columns and passing them to the model. But say we want to know the total cases for a particular day, the Random Forest Regressor (RFR) asks us to provide an entire row consisting of all other columns, as it takes every aspect into considration before predicting the outcome. 
Hence, we wish to know the most important features so that their values are fixed according to the date (w.r.t Linear Regression) and the values of other non-important or not so important features which barely change according to the date can be given the mean value of thier columns. 
Altogether, the row is constructed and passed to the regression model which then accepts it (as we are passing the perfect number of features) and predicts the outcome of our target.

Note: The order of this operation doesn't matter over here, i.e we can take the mean row first and overwrite the values of important features according to the date we select.  

```python
plt.figure(figsize=(12, 12))
# Create a pd.Series of features importances
importances = pd.Series(data=rf.feature_importances_,
                        index= X_train.columns)

# Sort importances
importances_sorted = importances.sort_values()

# Draw a horizontal barplot of importances_sorted
importances_sorted.plot(kind='barh', color='lightgreen')
plt.title('Features Importances')
plt.show()
```
